```python
#desordenamos los datos
data_condenados = data_condenados.sample(frac=1).reset_index(drop=True)
```

```python
def split_train_test(data,x_columns,y_column):
    #x_columns : array de columnas
    X = data_condenados[x_columns]
    y = data_condenados[y_column]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1237, stratify= y)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = split_train_test(data_condenados,features,'Reincidente')
```

```python
def Logistic_reg(data,features):
    X_train, X_test, y_train, y_test = split_train_test(data,features,'Reincidente')
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    logistic_regression = LogisticRegression()
    logistic_regression.fit(X_train_scaled, y_train)
    
    y_train_pred = logistic_regression.predict(X_train_scaled)
    y_test_pred = logistic_regression.predict(scaler.transform(X_test))
    accuracy_test = accuracy_score(y_test, y_test_pred)
    accuracy_train = accuracy_score(y_train, y_train_pred)

    return accuracy_train , accuracy_test
    #return X_train_scaled
```
